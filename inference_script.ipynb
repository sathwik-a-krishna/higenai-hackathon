```
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import torch
#from torch.utils.data import Dataset, DataLoader
#from datasets import load_dataset 
from transformers import AutoProcessor, BlipForConditionalGeneration
```

```
processor = AutoProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
model_trained = torch.load('models/model_highres.pt')
```

```
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
model_trained.to(device)
```

```
def generate_caption(model, image_path):
    inputs = processor(images=mpimg.imread(image_path), return_tensors="pt").to(device)
    pixel_values = inputs.pixel_values

    generated_ids = model.generate(pixel_values=pixel_values, max_length=50)
    generated_caption = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
    return generated_caption

```

```
# load image
image = 'dataset_highres/pic4.jpg'

print(generate_caption(model_trained, image))
```
